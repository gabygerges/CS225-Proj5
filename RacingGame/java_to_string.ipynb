```python
import os

def print_java_files_in_folder(folder_path):
    # List all files in the directory
    files = os.listdir(folder_path)
    
    # Filter out only .java files
    java_files = [f for f in files if f.endswith('.java')]
    
    # Sort the files alphabetically
    java_files.sort()
    
    # Iterate over each Java file
    for java_file in java_files:
        file_path = os.path.join(folder_path, java_file)
        
        # Print the header
        print(f"\n{'=' * 40}")
        print(f"File: {java_file}")
        print(f"{'=' * 40}\n")
        
        # Read and print the content of the file
        with open(file_path, 'r') as file:
            content = file.read()
            print(content)

# Example usage
folder_path = './'
print_java_files_in_folder(folder_path)
```
